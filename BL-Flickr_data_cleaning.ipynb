```python
import pandas as pd
import numpy as np 
from functools import reduce
```

# Dropping unnecessary columns

```python
data_frame = pd.read_csv('BL-Flickr-Images-Book.csv')
data_frame.head()
```

```python
column_drop = ['Edition Statement','Corporate Author','Corporate Contributors','Former owner','Engraver','Contributors',
           'Issuance type','Shelfmarks']
data_frame.drop(column_drop, inplace = True, axis = 1)
data_frame.head()
```

# Setting the index of the dataset

```python
data_frame.set_index('Identifier', inplace = True)
data_frame.head()
```

```python
data_frame['Date of Publication'].head(25)
```

```python
unwanted_characters = ['[', ',', '-']

def clean_dates(item):
    dop= str(item.loc['Date of Publication'])
    
    if dop == 'nan' or dop[0] == '[':
        return np.nan
    
    for character in unwanted_characters:
        if character in dop:
            character_index = dop.find(character)
            dop = dop[:character_index]
    
    return dop

data_frame['Date of Publication'] = data_frame.apply(clean_dates, axis = 1)
data_frame.head()
```

```python
def clean_author_names(author):
    
    author = str(author)
    
    if author == 'nan':
        return 'NaN'
    
    author = author.split(',')

    if len(author) == 1:
        name = filter(lambda x: x.isalpha(), author[0])
        return reduce(lambda x, y: x + y, name)
    
    last_name, first_name = author[0], author[1]

    first_name = first_name[:first_name.find('-')] if '-' in first_name else first_name
    
    if first_name.endswith(('.', '.|')):
        parts = first_name.split('.')
        
        if len(parts) > 1:
            first_occurence = first_name.find('.')
            final_occurence = first_name.find('.', first_occurence + 1)
            first_name = first_name[:final_occurence]
        else:
            first_name = first_name[:first_name.find('.')]
    
    last_name = last_name.capitalize()
    
    return f'{first_name} {last_name}'


data_frame['Author'] = data_frame['Author'].apply(clean_author_names)
data_frame.head()
```

```python
def clean_title(title):
    
    if title == 'nan':
        return 'NaN'
    
    if title[0] == '[':
        title = title[1: title.find(']')]
        
    if 'by' in title:
        title = title[:title.find('by')]
    elif 'By' in title:
        title = title[:title.find('By')]
        
    if '[' in title:
        title = title[:title.find('[')]

    title = title[:-2]
        
    title = list(map(str.capitalize, title.split()))
    return ' '.join(title)
    
data_frame['Title'] = data_frame['Title'].apply(clean_title)
data_frame.head()
```
