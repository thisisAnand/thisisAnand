## Renaming columns and skipping rows

```python
olympics_dataset = pd.read_csv('olympics.csv', skiprows = 1, header = 0)
olympics_dataset.head()
```

```python
new_names =  {'Unnamed: 0': 'Country',
              '? Summer': 'Summer Olympics',
              '01 !': 'Gold',
              '02 !': 'Silver',
              '03 !': 'Bronze',
              '? Winter': 'Winter Olympics',
              '01 !.1': 'Gold.1',
              '02 !.1': 'Silver.1',
              '03 !.1': 'Bronze.1',
              '? Games': '# Games', 
              '01 !.2': 'Gold.2',
              '02 !.2': 'Silver.2',
              '03 !.2': 'Bronze.2'}

olympics_dataset.rename(columns = new_names, inplace = True)
olympics_dataset.head()
```
